```python
import math

a = [15, 23, 88, 67, 35]

# 計算平均值
sum0 = 0
for e in a:
    sum0 += e
mean = sum0 / len(a)

# 計算變異數
sum1 = 0
for e in a:
    sum1 += (e - mean) ** 2
variance = sum1 / len(a)

# 計算標準差
std_dev = math.sqrt(variance)

print(std_dev)
```

```python
# 最大值

import math
import sys

a = [15, 23, 88, 67, 35]

max0 = -sys.maxsize - 1
for e in a:
    if e > max0:
        max0 = e

print(max0)
```

```python
# 最小值

import math
import sys
a = [15, 23, 88, 67, 35]

min0 = sys.maxsize
for e in a:
    if e < min0:
        min0 = e

print(min0)
```

```python
a = [15, 23, 88, 67, 35]

def total_of(a):
    sum0 = 0
    for e in a:
        sum0 += e
    return sum0

print(total_of(a))
```

```python
def search_key(a, key):
    index = -1
    for i in range(len(a)):
        if a[i] == key:
            index = i
            break
    return index

a = [15, 23, 88, 67, 35]
print(search_key(a, 67))
```

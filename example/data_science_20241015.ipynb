```python
a = ["老鼠", "老虎", "貓", "貓"] # list

def count_value(a, v):
    count = 0
    for i in a:
        if i == v:
            count += 1
    return count

v1 = count_value(a, "狗")
v2 = count_value(a, "貓")

print(v1)
print(v2)

# 計算 a 中的每個不重複元素出現幾次

def count_value(a):
    count = {}
    for i in a:
        if i not in count:
            count[i] = 1
        else:
            count[i] += 1
    return count

v3 = count_value(a)

print(v3)
```

```python
a = (1, 5, 8, 9) # tuple

def linear_search(a, key):
    for i in range(len(a)):
        if a[i] == key:
            return i, a[i]
    return -1, ''

b = linear_search(a, 8)

print(type(b))
print(b)
print(b[0], b[1])
```

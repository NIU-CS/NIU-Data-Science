```python
# 901

def main():
    with open("write.txt", "w", encoding="utf-8") as file:
        for i in range(5):
            data = input()
            file.write(data + "\n")

if __name__ == "__main__":
    main()
```

```python
# 902

def main():
    total = 0
    with open("read.txt", "r", encoding="utf-8") as file:
        content = file.read()
        numbers = map(int, content.split())
        total = sum(numbers)

    print(f"{total}")

if __name__ == "__main__":
    main()
```

```python
# 903

def main():
    with open("data.txt", "a", encoding="utf-8") as file:
        for i in range(5):
            name = input()
            file.write(name + "\n")

    with open("data.txt", "r", encoding="utf-8") as file:
        content = file.read()
        print(content)

if __name__ == "__main__":
    main()
```
